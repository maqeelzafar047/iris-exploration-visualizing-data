# **Step 01: Loading the Dataset**

### **There are several method of loading dataset:**

**1. Using a CSV File (with Pandas)**

You have downloaded the dataset as a **.csv** file from the internet **(e.g. UCI repository).**

Advantages:
* You can load any custom or real-world dataset
* Gives full control over file path, column names, and formatting
* Useful when working with external data or larger projects

**2. Using a Built-in Dataset (with Seaborn)**

You want to quickly explore and visualize a dataset **without downloading files.**

Advantages:
* Dataset is clean, preloaded, and easy to access

* Perfect for beginners or when you only need to analyze and visualize

* No file path or download needed â€” just **load by name**

```python
# Use method 1:
import pandas as pd

df = pd.read_csv('iris.csv')

df.head()
```

**Scatter Plot**

* This plot shows the relationship between Sepal Length and Sepal Width.
* It uses different colors (hue) to distinguish between different Iris species.
* Helps us understand how species are separated based on these two features

```python
# Scatter Plot:
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=df, x='SepalLengthCm', y='SepalWidthCm', hue='Species')
plt.title("Sepal Length vs Sepal Width by Species")
plt.show()
```

**Histograms of All Features**
* This block creates histograms for each numerical feature (excluding Id).
* It helps visualize the distribution (spread and shape) of each feature's values.
* Useful for understanding how data is distributed across measurements.


```python
# Histograms of All Features:
df.drop(columns=['Id']).hist(edgecolor='black', figsize=(10, 8))
plt.suptitle("Histograms of Iris Features")
plt.tight_layout()
plt.show()
```

**Histogram with KDE**
* This plot shows the distribution of Sepal Length only, with a KDE curve.
* KDE (Kernel Density Estimation) gives a smooth estimate of the distribution.
* Helps in seeing peaks, symmetry, and spread in a single variable.

```python
# Histogram with KDE:
sns.histplot(data=df, x='SepalLengthCm', kde=True)
plt.title("Distribution of Sepal Length")
plt.show()
```

**Box Plot**
* This plot shows Sepal Width values grouped by Species.
* Box plots display median, quartiles, and potential outliers.
* Useful for comparing distributions and spotting unusual values.

```python
# Box Plot:
sns.boxplot(data=df, x='Species', y='SepalWidthCm')
plt.title("Box Plot of Sepal Width by Species")
plt.show()
```
